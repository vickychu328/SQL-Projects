```python
#/* Q8: Produce a list of bookings on the day of 2012-09-14 which
#will cost the member (or guest) more than $30. Remember that guests have
#different costs to members (the listed costs are per half-hour 'slot'), and
#the guest user's ID is always 0. Include in your output the name of the
#facility, the name of the member formatted as a single column, and the cost.
#Order by descending cost, and do not use any subqueries. */

#Facilities Column names: ['facid', 'name', 'membercost', 'guestcost', 'initialoutlay', 'monthlymaintenance']
#Bookings Column names: ['bookid', 'facid', 'memid', 'starttime', 'slots']
#Members Column names: ['memid', 'surname', 'firstname', 'address', 'zipcode', 'telephone', 'recommendedby', 'joindate']


def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        
       SELECT Facilities.name AS facility_name, Members.firstname || ' ' || Members.surname AS member_name,Bookings.slots,
               CASE
                   WHEN Members.memid = 0 THEN Facilities.guestcost * Bookings.slots
                   ELSE Facilities.membercost * Bookings.slots
               END AS cost
        FROM Bookings
       JOIN Members ON Bookings.memid = Members.memid
       JOIN Facilities ON Bookings.facid= Facilities.facid
       WHERE Bookings.starttime LIKE'2012-09-14%'
        AND
        
        (
        Members.memid = 0 OR
        (Facilities.membercost * Bookings.slots) > 30
    )
        ORDER BY cost DESC
        
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

 
if __name__ == '__main__':
    main()
```

```python
# /* Q9: This time, produce the same result as in Q8, but using a subquery. */

#Facilities Column names: ['facid', 'name', 'membercost', 'guestcost', 'initialoutlay', 'monthlymaintenance']
#Bookings Column names: ['bookid', 'facid', 'memid', 'starttime', 'slots']
#Members Column names: ['memid', 'surname', 'firstname', 'address', 'zipcode', 'telephone', 'recommendedby', 'joindate']


def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        
SELECT
    facility_name,
    member_name,
    slots,
    cost
FROM (
    SELECT
        Facilities.name AS facility_name,
        Members.firstname || ' ' || Members.surname AS member_name,
        Bookings.slots,
        CASE
            WHEN Members.memid = 0 THEN Facilities.guestcost * Bookings.slots
            ELSE Facilities.membercost * Bookings.slots
        END AS cost,
        Bookings.starttime,
        Bookings.memid  -- Include memid in the subquery output
    FROM
        Bookings
    JOIN Members ON Bookings.memid = Members.memid
    JOIN Facilities ON Bookings.facid = Facilities.facid
) AS subquery
WHERE
    DATE(subquery.starttime) = '2012-09-14'
    AND (subquery.memid = 0 OR subquery.cost > 30)
ORDER BY
    cost DESC
    
    
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

 
if __name__ == '__main__':
    main()
```

```python
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        
SELECT Facilities.name,
    SUM(CASE
        WHEN Bookings.memid = 0 THEN Facilities.guestcost * Bookings.slots
        ELSE Facilities.membercost * Bookings.slots
    END)  AS total_revenue
FROM Facilities
JOIN Bookings on Bookings.facid = Facilities.facid
JOIN Members on Members.memid=Bookings.memid
GROUP BY
    Facilities.name
HAVING 
    total_revenue<1000
ORDER BY
    total_revenue;
    
    
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

 
if __name__ == '__main__':
    main()
```

```python
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
    
    
    SELECT
    facility_name,
    total_revenue
FROM (
    SELECT
        Facilities.name AS facility_name,
        SUM(CASE
            WHEN Bookings.memid = 0 THEN Facilities.guestcost * Bookings.slots
            ELSE Facilities.membercost * Bookings.slots
        END) AS total_revenue
    FROM
        Bookings
    JOIN Facilities ON Bookings.facid = Facilities.facid
    GROUP BY
        Facilities.name
) AS revenue_summary
WHERE
    total_revenue < 1000
ORDER BY
    total_revenue;
    
    
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

 
if __name__ == '__main__':
    main()
```

```python
# /* Q12: Find the facilities with their usage by member, but not guests */
#Facilities Column names: ['facid', 'name', 'membercost', 'guestcost', 'initialoutlay', 'monthlymaintenance']
#Bookings Column names: ['bookid', 'facid', 'memid', 'starttime', 'slots']
#Members Column names: ['memid', 'surname', 'firstname', 'address', 'zipcode', 'telephone', 'recommendedby', 'joindate']

def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
    
    
SELECT f.name AS facility_name, COUNT(b.bookid) AS useage_by_members
FROM Facilities f
JOIN Bookings b ON f.facid = b.facid
JOIN Members m ON b.memid = m.memid
WHERE m.memid != 0
GROUP BY f.name;

        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

 
if __name__ == '__main__':
    main()
    
```

```python
#/* Q13: Find the facilities usage by month, but not guests */
#Facilities Column names: ['facid', 'name', 'membercost', 'guestcost', 'initialoutlay', 'monthlymaintenance']
#Bookings Column names: ['bookid', 'facid', 'memid', 'starttime', 'slots']
#Members Column names: ['memid', 'surname', 'firstname', 'address', 'zipcode', 'telephone', 'recommendedby', 'joindate']

def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
    
    
    
SELECT f.name AS facility_name, strftime('%m',b.starttime)AS booking_month, COUNT(b.bookid) AS usage_by_members
FROM Facilities f
JOIN Bookings b ON f.facid = b.facid
JOIN Members m ON b.memid = m.memid
WHERE m.memid !=0
GROUP BY f.name, booking_month
ORDER BY f.name, booking_month;

        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

 
if __name__ == '__main__':
    main()
    
```
